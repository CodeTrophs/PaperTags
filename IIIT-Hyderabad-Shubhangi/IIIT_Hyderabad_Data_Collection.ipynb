## Importing Libraries

```
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

## Scrapping Data

```
main_page = urlopen('https://www.iiit.ac.in/people/faculty/')
main_soup = BeautifulSoup(main_page,'html.parser')
main_soup
```

### Getting Faculty Names

```
urls = []
faculty_names = []
td = main_soup.find_all('td')
for i in range(len(td)):
  divs = td[i].find_all('h3')
  for div in divs:
    if div.text != '\xa0':
      faculty_names.append(div.text)
      string = str(div)
      new_string = string.replace('<h3 class="name"><a href="','')
      new_string = new_string.replace(div.text,'')
      final_string = new_string.replace('"></a></h3>','')
      url = 'https://www.iiit.ac.in/' + final_string
      urls.append(url)
faculty_names
```

### Getting different Faculty's Page URL

```
urls
```

## Scrapping Each Faculty's Page for more information that is to find :
* Email ID
* Phone Number
* Research Areas

```
faculty_mail = []
faculty_phone = []
research_areas = []
for url in urls:
  faculty_page = urlopen(url)
  faculty_soup = BeautifulSoup(faculty_page,'html.parser')
  divs = faculty_soup.find_all('div',{'class':"container"})
  for div in divs:
    if div.text != '\xa0':
      string = str(div.text)
      faculty_details = string.strip().split('\n')
      print(faculty_details)
      index = faculty_details.index('Research Areas:')
      research_areas.append(faculty_details[index+1])
      for detail in faculty_details:
        if detail != '':
          if detail.startswith('Email'):
            faculty_mail.append(detail.strip().replace('Email: ',''))
          if detail.startswith('Phone: '):
            faculty_phone.append(detail.strip().replace('Phone: ',''))
```

```
research_areas
```

```
faculty_mail
```

```
final_faculty_mail = []
for mail in faculty_mail:
  if mail.find('[at]') != None:
    new_mail = mail.replace('[at]','@')
    new_mail = new_mail.replace('[dot]','.')
    final_faculty_mail.append(new_mail)
```

```
final_faculty_mail
```

```
faculty_phone
```

```
final_faculty_phone = []
for numbers in faculty_phone:
  index = numbers.find('Ext:')
  new_number = numbers[:index]
  final_faculty_phone.append(new_number)
```

```
final_faculty_phone
```

## Converting the collected data into pandas dataframe and finally transferring it to csv file

```
import pandas as pd
```

```
faculty = {'Faculty Name' : faculty_names, 'Email Id' : final_faculty_mail, 'Phone Number' : final_faculty_phone, 'Research Areas' : research_areas}
faculty
```

```
data = pd.DataFrame(faculty)
data
```

```
data.to_csv('IIIT_Hyderabad_Faculty_info.csv')
```
